```python
get_phase_error = lambda zernike: np.sqrt(np.nanmean(zernike**2))

nms = np.array([(2, 0), (2, -2), (2, 2), (3, -1), (3, 1)])
zern_names = ["Focus", "Astigmatism 45", "Astigmatism 0", "Coma X", "Coma Y"]
num_zernikes = 5


zernikes = []
zernikes_nan = []
for i in range(len(nms)):
    zernike_nan = get_zernike(nms[i, 0], nms[i, 1], pupil_size)
    norm_factor = 1/get_phase_error(zernike_nan)
    zernikes_nan.append(norm_factor * zernike_nan)
    zernikes.append(norm_factor * get_zernike(nms[i, 0], nms[i, 1], pupil_size, outside=0.))
    
zernikes = np.array(zernikes)
zernikes_nan = np.array(zernikes_nan)
```

```python
magnitude = np.abs(toliman_pupil)
toliman_phase = np.angle(toliman_pupil)
```

```python
aberrated_PSF_sample = []
aberrated_pupil_sample = []
for i in range(len(zernikes)):
    aberrated_pupil = get_wavefront(magnitude, toliman_phase + zernikes[i])
    aberrated_PSF = get_PSF(0, 0, aberrated_pupil, aperture, focal_length, true_wavelength, pixel_size, npixels, array_size)
    aberrated_pupil_sample.append(aberrated_pupil)
    aberrated_PSF_sample.append(aberrated_PSF)
```

## Bayes approach

```python
from jax import hessian
```

```python
def get_model(theta, toliman_pupil, phase_aberration, amplitude):
    """
    theta = [x, y]
    """
    aperture = 0.1
    focal_length = 3.65
    wavelength = 525e-9
    pixel_size = 2.5e-6
    array_size = 1024
    npixels = 512
    aberrated_pupil = get_wavefront(np.abs(toliman_pupil), np.angle(toliman_pupil) + (amplitude*phase_aberration))
    PSF = get_PSF(theta[0], theta[1], aberrated_pupil, aperture, focal_length, true_wavelength, pixel_size, npixels, array_size)
    return PSF

def nlp(theta, data, noise, toliman_pupil, phase_aberration, amplitude):
    """
    Negative log prior - chi squared
    """
    model = get_model(theta, toliman_pupil, phase_aberration, 0.)
    norm_diff =  (data - model)/noise
    chi2 = np.sum(norm_diff**2)
    return -chi2/2
nlp_hessian = hessian(nlp, argnums=0)

def cov_mat(theta, noise, toliman_pupil, phase_aberration, amplitude):
    """
    Covariance matrix
    """
    data = get_model(theta, toliman_pupil, phase_aberration, amplitude)
    H = nlp_hessian(theta, data, noise, toliman_pupil, phase_aberration, amplitude)
    return np.linalg.inv(H)
```

```python
amplitudes = np.linspace(0, 1, 20, endpoint=False)
print(amplitudes)
```

```python
theta = np.array([1., 0.,])
values = []
for i in tqdm(range(len(zernikes))):
    values_temp = []
    for j in range(len(amplitudes)):
        sigma = cov_mat(theta, 1, toliman_pupil, zernikes[i], amplitudes[j])
        values_temp.append(np.mean(np.sqrt(np.diag(np.abs(sigma)))))
    values.append(values_temp)
```

```python
values = np.array(values)
```

```python
colours = ["tab:blue", "tab:orange", "tab:orange", "tab:green", "tab:green"]
linestyles = ["solid", "solid", "dotted", "solid", "dotted"]
```

```python
plt.figure(figsize=(8, 4))
for i in range(len(zernikes)):
    plt.plot(amplitudes, values[i, :], label=zern_names[i], c=colours[i], linestyle=linestyles[i])
    
plt.xlabel("Aberration Amplitude")
plt.ylabel("Parameter Constraint")
plt.legend()
plt.tight_layout()
plt.savefig("Bayes_analysis")
plt.show()
```
