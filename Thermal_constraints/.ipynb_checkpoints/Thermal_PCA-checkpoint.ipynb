```python
aperture = 0.1
focal_length = 3.65
wavelength = 525e-9
pixel_size = 2.5e-6
array_size = 1024
# npixels = 512
npixels = 256
```

```python
pupil_size, true_wavelength = get_wavefront_size(aperture, focal_length, wavelength, pixel_size, array_size)
print(pupil_size, true_wavelength)
print("Samlping is {:.2f}x Nyquist".format(array_size/(2*pupil_size)))
toliman_pupil = get_toliman_pupil(pupil_size)
```

```python
toliman_PSF = get_PSF(0, 0, toliman_pupil, aperture, focal_length, true_wavelength, pixel_size, npixels, array_size)
```

```python
get_phase_error = lambda zernike: np.sqrt(np.nanmean(zernike**2))

nms = np.array([(2, 0), (2, -2), (2, 2), (3, -1), (3, 1)])
zern_names = ["Focus", "Astigmatism 45", "Astigmatism 0", "Coma Y", "Coma Y"]
num_zernikes = 5


zernikes = []
zernikes_nan = []
for i in range(len(nms)):
    zernike_nan = get_zernike(nms[i, 0], nms[i, 1], pupil_size)
    norm_factor = 1/get_phase_error(zernike_nan)
    zernikes_nan.append(norm_factor * zernike_nan)
    zernikes.append(norm_factor * get_zernike(nms[i, 0], nms[i, 1], pupil_size, outside=0.))
    
zernikes = np.array(zernikes)
zernikes_nan = np.array(zernikes_nan)
```

```python
magnitude = np.abs(toliman_pupil)
toliman_phase = np.angle(toliman_pupil)
```

```python
aberrated_PSF_sample = []
aberrated_pupil_sample = []
for i in range(len(zernikes)):
    aberrated_pupil = get_wavefront(magnitude, toliman_phase + zernikes[i])
    aberrated_PSF = get_PSF(0, 0, aberrated_pupil, aperture, focal_length, true_wavelength, pixel_size, npixels, array_size)
    aberrated_pupil_sample.append(aberrated_pupil)
    aberrated_PSF_sample.append(aberrated_PSF)
```

```python
plt.figure(figsize=(25, 4))
for i in range(len(zernikes)):
    plt.subplot(1, 5, i+1)
    plt.imshow(zernikes[i])
    plt.title(zern_names[i])
    plt.colorbar()
plt.show()

plt.figure(figsize=(25, 4))
for i in range(len(zernikes)):
    plt.subplot(1, 5, i+1)
    plt.imshow(np.angle(aberrated_pupil_sample[i]))
    plt.title(zern_names[i])
    plt.colorbar()
plt.show()

plt.figure(figsize=(25, 4))
for i in range(len(zernikes)):
    plt.subplot(1, 5, i+1)
    plt.imshow(aberrated_PSF_sample[i])
    plt.title(zern_names[i])
    plt.colorbar()
plt.show()
```

# PCA method

```python
from sklearn.decomposition import PCA
```

```python
namplitudes = 100
max_amplitude = 1
amplitudes = np.linspace(0, max_amplitude, namplitudes, endpoint=False)
```

```python
aberrated_PSFs = []

for i in tqdm(range(len(zernikes))):
    aberrated_PSFs_temp = []
    for j in tqdm(range(len(amplitudes)), leave=False):
        aberrated_pupil = get_wavefront(magnitude, toliman_phase + (amplitudes[j] * zernikes[i]))
        aberrated_PSFs_temp.append(get_PSF(0, 0, aberrated_pupil, aperture, focal_length, true_wavelength, pixel_size, npixels, array_size))
    aberrated_PSFs.append(aberrated_PSFs_temp)
    
aberrated_PSFs = np.array(aberrated_PSFs)
```

```python
variances = np.zeros([num_zernikes, namplitudes, namplitudes])
for i in tqdm(range(num_zernikes)):
    for j in tqdm(range(1, namplitudes), leave=False):
        data = aberrated_PSFs[i, :j+1].reshape([j+1, npixels**2])
        pca = PCA().fit(data)
        var = np.array(pca.explained_variance_ratio_)
        variances = index_update(variances, index[i, j, :j+1], var)
```

```python
variances_all = np.zeros([namplitudes, num_zernikes*namplitudes])

for j in tqdm(range(1, namplitudes), leave=False):
    data = aberrated_PSFs[:, :j+1].reshape([num_zernikes*(j+1), npixels**2])
    pca = PCA().fit(data)
    var = np.array(pca.explained_variance_ratio_)
    variances_all = index_update(variances_all, index[j, :num_zernikes*(j+1)], var)
```

```python
nvectors = 10
Xs = amplitudes
Ys = np.arange(0, nvectors)
X, Y = np.meshgrid(Xs, Ys)
Z = np.log10(variances)

nvectors_all = 50
Xs_all = amplitudes
Ys_all = np.arange(0, nvectors_all)
X_all, Y_all = np.meshgrid(Xs_all, Ys_all)
Z_all = np.log10(variances_all)
```

```python
mask1 = Z_all < -35
mask2 = Z_all != -np.inf
mask = mask1*mask2
Z_all = index_update(Z_all, index[mask], -35)
```

```python
plt.figure(figsize=(18, 15))

plt.subplot(3, 2, 1)
plt.contourf(X_all, Y_all, Z_all[:, :nvectors_all].T, levels=6)
plt.title("All Zernikes")
plt.xlabel("Zernike Amplitude (RMS radians)")
plt.ylabel("Number of Basis vectors")
plt.colorbar()

for i in range(num_zernikes):
    plt.subplot(3, 2, i+2)
    plt.contourf(X, Y, Z[i, :, :nvectors].T, levels=6)
    plt.title(zern_names[i])
    plt.xlabel("Zernike Amplitude (RMS radians)")
    plt.ylabel("Number of Basis vectors")
    plt.colorbar()

plt.tight_layout()
plt.savefig("PCA_result")
plt.show()
```
