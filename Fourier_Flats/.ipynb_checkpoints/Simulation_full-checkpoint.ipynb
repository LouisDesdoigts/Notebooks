```python
# def propagate_wavefront(wavefront):
def propagate_wavefront_min(wavefront):
    """
    Uses Fourier Transforms to propagate a wavefront from the aperture (pupil plane) to the focal plane.
    Parameters: 
        wavefront: a square ([n, n]) complex array representing the wavefront at the telescope aperture.
    Returns:
        a square ([n, n]) complex array which represents the wavefront at the focal plane.
    """
    return np.fft.fftshift(np.fft.fft2(wavefront))

# def get_wavefront(magnitudes, phases):
def get_wavefront_min(magnitudes, phases):
    """
    Combines the real and magnitude and phase arrays into a complex array representing the wavefront.
    Parameters: 
        magnitudes: a square ([n, n]) array representing the telescope aperture, typically a top-hat function.
        phases: a square ([n, n]) array representing the phase of the light at the aperture, this will typically encode the TOLIMAN pupil.
    Returns:
        wavefront: a square ([n, n]) complex array representing the wavefront at the telescope aperture.
    """
    wavefront = magnitudes * np.exp(1j * phases)
    return wavefront

# def get_intensity(wavefront):
def get_intensity_min(wavefront):
    """
    Takes some complex array and returns its intensity (modulus squared)
    Parameters:
        wavefront: a square ([n, n]) complex array
    Returns: 
        intensity: The intensity of the input array (ie its modulus squared, ([n, n]))
    """
    intensity = np.abs(wavefront)**2
    return intensity

# def calc_delta_phase(x, y, aperture, focal_length, wavelength):
def calc_delta_phase_min(x, y, aperture, focal_length, wavelength):
    """
    Takes the optical parameters of the system and calcualtes the change in phase 
    across the aperture required to shift the PSF by x, y
    Parameters:
        x, y: The x and y position on the chip (m)
        aperture: Diameter of the aperture (m)
        focal_length: Focal length (m)        
        wavelength: Wavelength of light beng similated (m)
    Returns:
        delta_phase: The total change in phase across the aperture that shifts the
        PSF by x, y
    """
#     # Total radial deviation from the centre of the chip
#     r0 = np.hypot(x, y)

#     # Paths lengths of each light arriving from opposite sides of the aperture
#     R = aperture/2
#     l1 = np.hypot((R + r0)/wavelength, (focal_length)/wavelength)
#     l2 = np.hypot((R - r0)/wavelength, (focal_length)/wavelength)

#     # Difference in path lengths and phase across the aperture
#     delta_length = np.abs(l1 - l2)

#     delta_phase = 2 * np.pi * delta_length

    # Total radial deviation from the centre of the chip
    r0 = np.hypot(x, y)

    # Paths length difference
    delta_length = aperture * r0 / np.hypot(focal_length, r0)

    # Total phase change
    delta_phase = 2 * np.pi * delta_length / wavelength
    
    return delta_phase

# def get_phase_slope(x_pix, y_pix, aperture, focal_length, wavelength, pixel_size, array_size):
def get_phase_slope_min(x_pix, y_pix, aperture, focal_length, wavelength, pixel_size, array_size):
    """
    Creates an array of phases at the aperture which is required to shift the PSF by x, y
    Parameters:
        x_pix, y_pix: x and y position on the detector (pixels)
        aperture: Diameter of the aperture (m)
        focal_length: Focal length (m)        
        wavelength: @avelength of light beng similated (m)
        pixel_size: Linear dimension of a single pixel (m)
        array_size: Size of the output array
    Returns:
        phase_slope: An array of phases which shifts the PSF by x_pix, y_pix
    """
    # Convert x, y from pixels into m
    x = x_pix * pixel_size
    y = y_pix * pixel_size
    
    # Calculate the total phase change
    delta_phase = calc_delta_phase(x, y, aperture, focal_length, wavelength)
    
    # Generate array of phases
    phases = np.linspace(1, 1+delta_phase, array_size, endpoint=False)
    # Note the extra 1 is there to allow differentiablity of the function
    
    # Convert the array of phases into polar coordinates
    Xs, Ys = np.meshgrid(phases, phases)
    Rs = np.hypot(Xs, Ys)
    Phis = np.arctan2(Ys, Xs)
    
    # Calculate the rotation angle
    theta = np.arctan2(y, x)
    
    # Reconstruct the phases in cartesian coordinates while applying the rotation angle
    phase_slope = Rs * np.cos(Phis + theta)
    
    return phase_slope
    
# def calc_padding(aperture, focal_length, wavelength, pixel_size):
def calc_padding_min(aperture, focal_length, wavelength, pixel_size):
    """
    Calcualtes the padding required in order to correctly sample frequencies in the focal plane
    Parameters:
        aperture: Diameter of the aperture (m)
        focal_length: Focal length (m)        
        wavelength: Wavelength of light beng similated (m)
        pixel_size: Linear dimension of a single pixel (m)
    Returns: 
        padding: The padding factor required for the optical system
        
    Note: 
        This prints a warning to let people know that such an optical system is undersampled, ie below Nyquist
        @Peter, where does this equation come from?
    """
    plate_scale = np.arctan2(pixel_size, focal_length)    # Radians per pixel
    spatial_freq = wavelength/aperture
    padding = spatial_freq/plate_scale
        
    return padding

# def get_wavefront_size(aperture, focal_length, wavelength, pixel_size, array_size):
def get_wavefront_size_min(aperture, focal_length, wavelength, pixel_size, array_size):
    """
    Takes the optical parameters and calculates the size of the array representing the 
    wavefront must be. It also returns what true wavelength this represents.
    Note wavefront_size is returned as a native python integer rather than 
    Parameters:
        aperture: Diameter of the aperture (m)
        focal_length: Focal length (m)        
        wavelength: Wavelength of light beng similated (m)
        pixel_size: Linear dimension of a single pixel (m)
        array_size: The linear dimension of the array to be transformed 
    Returns:
        wavefront_size: The size of the required wavefront array to simulate the desired wavelength
        true_wavelength: The true wavelength that is being simulated given the parameters (m)
    """
    padding = calc_padding(aperture, focal_length, wavelength, pixel_size)
    wavefront_size = np.round(array_size/padding, decimals=0).astype(np.int16)
    true_wavelength = array_size * np.arctan2(pixel_size, focal_length) * aperture / wavefront_size
#     return int(wavefront_size), true_wavelength
    return wavefront_size, true_wavelength

# def get_PSF(x_pix, y_pix, pupil, aperture, focal_length, wavelength, pixel_size, npixels, array_size):
def get_PSF_min(x_pix, y_pix, pupil, aperture, focal_length, wavelength, pixel_size, npixels, array_size):
    """
    Generates a monochromatic PSF of the optical system centered about the point (x_pix, ypix).
    Note that the normalisation of the PSF is done before the region of the array that represents 
    the detector is cut out, therefore when npixels < array_size, sum(PSF) < 1.
    Parameters:
        x_pix: The x position of the PSF (pix)
        y_pix: The y position of the PSF (pix)
        pupil: A complex array representing both the amplitude and phase at the aperture
        of the telescope.
        aperture: Diameter of the aperture (m)
        focal_length: Focal length (m)        
        wavelength: Wavelength of light beng similated (m)
        pixel_size: Linear dimension of a single pixel (m)
        npixels: The linear dimension of the detector (pix)
        array_size: The linear dimension of the array to be transformed 
    Returns:
        PSF: The normalised PSF of the optical system
    """
    
    # Get phase slope
    pupil_size = pupil.shape[0]
    phase_slope = get_phase_slope(x_pix, y_pix, aperture, focal_length, wavelength, pixel_size, pupil_size)
    
    # Generate inital wavefront
    aperture_magnitude = np.abs(pupil)
    aperture_phases = np.angle(pupil) + phase_slope
    initial_wavefront = get_wavefront(aperture_magnitude, aperture_phases)
    
    # Pad array for transforming
    zeros_array = np.zeros([array_size, array_size], dtype=np.complexfloating)
    print(zeros_array)
    padded_wavefront = index_update(zeros_array, index[:pupil_size, :pupil_size], initial_wavefront)
    
    # Propagate the wavefront and get the PSF
    final_wavefront = propagate_wavefront(padded_wavefront)
    PSF_full = get_intensity(final_wavefront)
    PSF_norm = PSF_full/np.sum(PSF_full)
    
    # Take region of chip we care about
    c = array_size//2
    s = npixels//2
    PSF = PSF_norm[c-s:c+s, c-s:c+s]
    
    return PSF

def get_empty_pupil(pupil_size):
    """
    Generates a complex array to represent an empty telescope aperture
    (ie one that generates an airy disk)
    Parameters:
        pupil_size: The linear dimension of the desired pupil (pix)
    Return:
        pupil: A complex array representing the pupil of an empty telescope aperture
    """
    magnitude = get_aperture(pupil_size)
    phases = np.zeros([pupil_size, pupil_size])
    pupil = get_wavefront(magnitude, phases)
    return pupil

def get_aperture(size, radius=None):
    """
    Produces an array to be used to represent the magnitude of light at the aperture of a telescope.
    Ie an array of zeros with a circle of ones at the centre
    Parameters:
        size: The size of the array to return
        radius (optional): The radius which defines the boundary of the aperture, if left as None it
        will extend the aperture to the edged of the array (ie radius = size/2)
    Returns:
        aperture: An array of zeros with a circular section in the middle of ones
    """
    c = size/2
    xs = np.linspace(-c, c, size, endpoint=False)
    Xs, Ys = np.meshgrid(xs, xs)
    Rs = np.hypot(Xs, Ys)
    
    if radius is None:
        aperture = np.where(Rs > c, 0, np.ones([size, size]))
    else:
        aperture = np.where(Rs > radius, 0, np.ones([size, size]))
    return aperture
```

```python
propagate_wavefront = jit(propagate_wavefront_min)
get_wavefront = jit(get_wavefront_min)
get_intensity = jit(get_intensity_min)
calc_delta_phase = jit(calc_delta_phase_min)
get_phase_slope = jit(get_phase_slope_min, static_argnums=(6))
calc_padding = jit(calc_padding_min)
get_wavefront_size = jit(get_wavefront_size_min)
get_PSF = jit(get_PSF_min, static_argnums=(7, 8))
```

### Random Fucntions

```python
def make_mask(pupil_size, array_size, npix, r=0):
    xs = np.arange(-npix//2, npix//2)
    XX, YY = np.meshgrid(xs, xs)
    RR = np.hypot(XX, YY)
    mask = RR > pupil_size/(array_size/npix) + r
    return mask

def square_super_gaussian(sigma, power, npix):
    c = npix/2
    xs = np.linspace(-c, c, npix, endpoint=False)
    exponent = ((xs**2) / (2*sigma**2))**power
    Gx = np.tile(np.exp(-exponent), npix).reshape(npix, npix)
    Gy = np.repeat(np.exp(-exponent), npix).reshape(npix, npix)
    return Gx*Gy

def super_gaussian(sigma, power, npix):
    c = npix/2
    xs = np.linspace(-c, c, npix, endpoint=False)
    XX, YY = np.meshgrid(xs, xs)
    
    exponent = (0.5*( (XX/sigma)**2 + (YY/sigma)**2))**power
    G = np.exp(-exponent)
    return G
```

```python
def get_power_spectrum_min(image):
    return np.abs(np.fft.fftshift(np.fft.fft2(image)))**2
get_power_spectrum = jit(get_power_spectrum_min)

def ADAM(X, grad, LR, i, m, v, b1=0.9, b2=0.999, epsilon=1):
    m = b1 * m + (1 - b1) * grad
    v = b2 * v + (1 - b2) * np.power(grad, 2)
    m_hat = m / (1 - np.power(b1, i+1))
    v_hat = v / (1 - np.power(b2, i+1))
    X_new = X - LR * m_hat / (np.sqrt(v_hat) + epsilon)
    return X_new, m, v

def error(X, images, mask):
    # Mean shift
    X -= np.mean(X) - 1
    
    # Apply FF correction to images
    Xsq = X.reshape(images.shape[-2:])
    corrected_images = images * Xsq
    
    # Get the power spectrum of all the images
    images_power_spec = get_power_spectrum(corrected_images)
    images_power_spec /= np.sum(images_power_spec)
    
    # Get sum of frequncies above the diffraction limit
    power_spec_outer_sum = np.sum(mask * images_power_spec)

    return power_spec_outer_sum
err_grad = value_and_grad(error, argnums=0)

def ent_error(X, images, mask, alpha=1):
    
    mean_image = np.mean(images, axis=0)
    mean_image /= np.max(mean_image)
    flux_mask = (mean_image > 0.25).flatten()
    
    # Mean shift
    X -= np.mean(X) - 1
    
    # Apply FF correction to images
    Xsq = X.reshape(images.shape[-2:])
    corrected_images = images * Xsq
    
    # Get the power spectrum of all the images
    images_power_spec = get_power_spectrum(corrected_images)
    images_power_spec /= np.sum(images_power_spec)
    
    # Get sum of frequncies above the diffraction limit
    power_spec_outer_sum = np.sum(mask * images_power_spec)
    
    # Get entropy
    entropy = alpha*get_entropy(1/X[flux_mask])

    return power_spec_outer_sum - entropy
ent_err_grad = value_and_grad(ent_error, argnums=0)

def gauss_error(X, images, mask):
    
    # Create Gaussian
    G = super_gaussian(X[0], X[1], images.shape[-1])
    corrected_images = G * images
    
    # Get the power spectrum of all the images
    images_power_spec = get_power_spectrum(corrected_images)
    images_power_spec /= np.sum(images_power_spec)
    
    # Get sum of frequncies above the diffraction limit
    power_spec_outer_sum = np.sum(mask * images_power_spec)

    return power_spec_outer_sum
gauss_err_grad = value_and_grad(gauss_error, argnums=0)
```

```python
# def sigmoid(xs):
#     xs -= np.mean(xs) - 0.5 # Normalise to 0.5
#     s = 1/(1 + np.exp(-xs))
#     return s

def get_entropy(xs, axis_in=(-1)):
#     p = sigmoid(xs)
    p = xs
    p /= np.sum(p)
    ent = -np.sum(p*np.log(p), axis=axis_in)
    return ent

def get_RMSE(FF_found, FF_true, axis_in=(-1)):
    return np.sqrt(np.mean(np.square(FF_found - FF_true), axis=axis_in))

# def get_WRMSE(FF_found, FF_true, mean_image):
#     return np.sqrt(np.mean(mean_image*(np.square(FF_found - FF_true))))
```

```python
# Uniform Initilatisation
X = np.ones(npix**2)
alpha = 0

# Random Initilisation (matching true parameters)
# X = np.array(onp.random.normal(1, 0.05, npix**2))

# Find inital LR
LR = 1e10
while True:
#     err0, grad = err_grad(X, data, mask)
    err0, grad = ent_err_grad(X, data, mask, alpha)
    m, v = 0, 0
    for i in range(5):
        X_new, m, v = ADAM(X, grad, LR, i, m, v)
#         err1, grad = err_grad(X_new, data, mask)
        err1, grad = ent_err_grad(X_new, data, mask, alpha)
    if err0 <= err1 or np.isnan(err1):
        LR *= 0.1
    else:
        print("log_10 LR in: {}".format(np.log10(LR)))
        break

# Descend Gradients
m, v = 0, 0
errs = []
grads = []
Xs = [X]
epochs = [500]


for k in tqdm(range(epochs[-1])):
    # Get gradients and update X
#     err, grad = err_grad(X, data, mask)
    err, grad = ent_err_grad(X, data, mask, alpha)
    X, m, v = ADAM(X, grad, LR, k, m, v)
    
    # Append things
    Xs.append(X)
    errs.append(err)
    grads.append(grad)
    
Xs = np.array(Xs)
errs = np.array(errs)
grads = np.array(grads)
```
